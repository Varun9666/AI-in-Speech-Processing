```python
#A1

import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import IPython.display as ipd

audio_path = 'Varun.wav'
y, sr = librosa.load(audio_path)

derivative_1 = np.diff(y)
derivative_1 /= np.max(np.abs(derivative_1))

print("Playing First Derivative Signal:")
plt.figure(figsize=(10, 5))
librosa.display.waveshow(derivative_1)
plt.title('First Derivative of Audio Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.tight_layout()
plt.show()

ipd.Audio(derivative_1, rate=sr)
```

```python
#Original
plt.figure(figsize=(10, 5))
librosa.display.waveshow(y, sr=sr)
plt.title('ORIGINAL')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.tight_layout()
plt.show()

ipd.Audio(y, rate=sr)
```

```python
#A2

zero_crossings = np.where(np.diff(np.sign(derivative_1)))[0]
differences = np.diff(zero_crossings)
threshold = 1000
speech_regions = differences[differences > threshold]
silence_regions = differences[differences <= threshold]

average_length_speech = np.mean(speech_regions)
average_length_silence = np.mean(silence_regions)

print("Average length between consecutive zero crossings in speech regions:", average_length_speech)
print("Average length between consecutive zero crossings in silence regions:", average_length_silence)

plt.figure(figsize=(10, 5))
plt.plot(differences, label='All regions', color='black')
plt.plot(np.arange(len(speech_regions)), speech_regions, 'ro', label='Speech regions', color='blue')
plt.plot(np.arange(len(speech_regions), len(speech_regions) + len(silence_regions)), silence_regions, 'bo', label='Silence regions', color='blue')
plt.title('Pattern of Zero Crossings')
plt.xlabel('Zero Crossing Difference')
plt.ylabel('Difference between Consecutive Zero Crossings')
plt.legend()
plt.show()

print("Pattern of Zero Crossings:")
print("All regions -", differences)
print("Speech regions -", speech_regions)
print("Silence regions -", silence_regions)
```

```python
#A3

my_word_files = ['Paytm.wav', 'Amazon.wav', 'Apple.wav', 'Microsoft.wav', 'Tesla.wav']
teammate_word_files = ['Paytm_1.wav', 'Amazon_1.wav', 'Apple_1.wav', 'Microsoft_1.wav', 'Tesla_1.wav']
words = ['Paytm', 'Amazon', 'Apple', 'Microsoft', 'Tesla']
my_word_lengths = []
teammate_word_lengths = []

for my_word_file in my_word_files:
    my_signal, my_sr = librosa.load(my_word_file, sr=None)
    my_length_seconds = len(my_signal) / my_sr
    my_word_lengths.append(my_length_seconds)

for teammate_word_file in teammate_word_files:
    teammate_signal, teammate_sr = librosa.load(teammate_word_file, sr=None)
    teammate_length_seconds = len(teammate_signal) / teammate_sr
    teammate_word_lengths.append(teammate_length_seconds)

print("Lengths of the spoken words - My Recordings - ", my_word_lengths)
print("Lengths of the spoken words - Teammate's Recordings - ", teammate_word_lengths)

bar_width = 0.5
index = np.arange(len(words))
plt.figure(figsize=(12, 6))
plt.bar(index - bar_width/2, my_word_lengths, bar_width, label='My Words', color='lightgreen')
plt.bar(index + bar_width/2, teammate_word_lengths, bar_width, label="Teammate's Words", color='salmon')
plt.xlabel('Words')
plt.ylabel('Length (seconds)')
plt.title('Comparison of Spoken Words Length')
plt.xticks(index, words)
plt.legend()

plt.show()
```
